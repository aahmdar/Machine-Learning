```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_csv('https://drive.google.com/u/0/uc?id=1f_lwmsFZ8UP0XkMeI45RtM4cpZrKcDTX&export=download',header=None)
df.rename({0:'Date'},axis=1,inplace=True)
df
```

```python
df[3]=1
data_gr = (df.groupby([1,2])[3].sum().unstack().reset_index().fillna(0) .set_index(1))

def encode_units(x): 
  if x <= 0: return 0 
  if x >= 1: return 1 
data_sets = data_gr.applymap(encode_units) 
data_sets
```

```python
ap = apriori(data_sets, min_support=0.3, use_colnames=True)

ar = association_rules(ap, metric="lift", min_threshold=1) # create association rules
# rules.loc[rules['lift'] == rules['lift'].max()] # select rules with the highest lift
ar
```
